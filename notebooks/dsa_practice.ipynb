```python
from collections import Counter
import heapq

def topKFrequent(nums, k):
    freq = Counter(nums)
    heap = []

    for num, count in freq.items():
        heapq.heappush(heap, (count, num))
        print("heap push")
        print(heap)
        if len(heap) > k:
            heapq.heappop(heap)
            print("heap pop")
            print(heap)

    print(heap)

    return [num for count, num in heap]


nums = [1,1,1,2,2,3]
k = 2

print(topKFrequent(nums, k))

```

```python
def missingNumber(nums):
    xor = len(nums)
    for i, num in enumerate(nums):
        # print(num)
        print(f"{str(i)} -> {str(num)}" )
        xor ^= i ^ num
        print(xor)
    return xor


print(missingNumber([1,3,0]))
```

```python

#prefix = [1,1,2,6]
#postfix = [24,12,4,1]
#res = [24,12,8,6]

def productExceptSelf(nums):
    res = [1]*len(nums)
    prefix = 1
    postfix = 1
    for i in range(len(nums)):
       res[i] = prefix
       prefix *= nums[i]

    for i in reversed(range(len(nums))):
        res[i] *= postfix
        postfix *= nums[i]

    print(res)



nums = [1,2,3,4]
productExceptSelf(nums)
```
