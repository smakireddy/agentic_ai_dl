```python
from collections import Counter
import heapq
from typing import List
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # frequency = {}
        # for num in nums:
        #     if num in frequency.keys():
        #         frequency[num] +=1
        #     else:
        #         frequency[num] = 1
        # # result = dict(sorted(frequency.items(), key=lambda x:x[1], reverse=True)[:k])
        # result = [num for num,_ in sorted(frequency.items(), key=lambda x:x[1], reverse=True)[:k]]

        # return result
        frequency = Counter(nums)
        heap = []
        for num,count in frequency.items():
            # print(heap)
            heapq.heappush(heap, (count, num))
            # heapq.heappush(heap,(num,count))
            # print(heap)
            if len(heap) > k:
                # print(heap)
                heapq.heappop(heap)
                # print(heap)

        return [num for count, num in heap]

```
