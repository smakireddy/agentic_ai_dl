```python
from collections import defaultdict

scores = [
    ["Bobby", "87"],
    ["Charles", "100"],
    ["Eric", "64"],
    ["Charles", "22"]
]

student_scores = defaultdict(list)

for name, marks in scores:
    student_scores[name].append(int(marks))

print(dict(student_scores))
max_avg = -1
max_avg_student = None
for k,v in student_scores.items():
    avg_marks =  sum(v)/len(v)
    if avg_marks > max_avg:
        max_avg = avg_marks
        max_avg_student = k
print(max_avg_student)

print("Alternative.........")
print(max(student_scores, key = lambda x : sum(student_scores[x])/len(student_scores[x])))


```

```python
posts = [ (1, '2024-01-01 08:00:00'),
          (2, '2024-01-01 09:15:00'),
          (1, '2024-01-01 10:00:00'),
          (3, '2024-01-01 11:45:00'),
          (3, '2024-01-01 12:00:00'), ]
user_count = defaultdict(int)
for user_id, _ in posts :
    user_count[user_id] += 1

print(user_count)
```

```python
from datetime import datetime

def get_hour(ts):
    return datetime.strptime(ts, '%Y-%m-%d %H:%M:%S').hour

print (get_hour("2024-01-01 08:00:00"))
```
